```python
import torch
from transformers import GPT2TokenizerFast, GPT2LMHeadModel
import nltk
from nltk.corpus import wordnet as wn
import torch.nn.functional as F



# Function to get GPT-2 embedding for a word
def get_word_embedding(word):
    input_ids = tokenizer([word], return_tensors="pt", add_special_tokens=False)['input_ids']
    with torch.no_grad():
        embedding = model.transformer.wte(input_ids).mean(dim=1)  # Average over token embedding
    return embedding

# Function to compute cosine similarity
def cosine_similarity(embedding1, embedding2):
    embedding1_norm = F.normalize(embedding1, p=2, dim=1)
    embedding2_norm = F.normalize(embedding2, p=2, dim=1)
    return torch.matmul(embedding1_norm, embedding2_norm.T).item()

# Function to check if two words are synonyms using WordNet
def are_synonyms(word1, word2):
    synsets1 = wn.synsets(word1)
    synsets2 = wn.synsets(word2)

    for synset1 in synsets1:
        for synset2 in synsets2:
            if synset1 == synset2:
                return True
    return False

import torch
from transformers import GPT2TokenizerFast, GPT2LMHeadModel
from nltk.corpus import wordnet as wn
import torch.nn.functional as F
import random


tokenizer = GPT2TokenizerFast.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')
model.eval()

def get_word_embedding(word):
    input_ids = tokenizer([word], return_tensors="pt", add_special_tokens=False)['input_ids']
    with torch.no_grad():
        embedding = model.transformer.wte(input_ids).mean(dim=1)  # Average over token embedding
    return embedding

def cosine_similarity(embedding1, embedding2):
    embedding1_norm = F.normalize(embedding1, p=2, dim=1)
    embedding2_norm = F.normalize(embedding2, p=2, dim=1)
    return torch.matmul(embedding1_norm, embedding2_norm.T).item()

# Function to check if two words are synonyms using WordNet
def are_synonyms(word1, word2):
    synsets1 = wn.synsets(word1)
    synsets2 = wn.synsets(word2)

    for synset1 in synsets1:
        for synset2 in synsets2:
            if synset1 == synset2:
                return True
    return False

def add_ing(verb):
    """
    Adds 'ing' to a given verb following English spelling rules.
    """
    if verb.endswith('ie'):  # Change 'ie' to 'y' and add 'ing'
        return verb[:-2] + 'ying'
    elif verb.endswith('e') and not verb.endswith('ee'):  # Drop 'e' (except for 'ee') and add 'ing'
        return verb[:-1] + 'ing'
    elif verb[-1] in 'bcdfghjklmnpqrstvwxyz' and verb[-2] in 'aeiou' and len(verb) > 1:  # Check if ends with a consonant after a vowel
        if (len(verb) == 2 or (verb[-3] not in 'aeiou' and verb[-2] == verb[-1])):  # Check if it's a short vowel or the final syllable is stressed
            return verb + verb[-1] + 'ing'  # Double the final consonant if short vowel and add 'ing'
        else:
            return verb + 'ing'
    elif verb.endswith('c'):  # Ends with 'c', add 'k' before 'ing'
        return verb + 'king'
    else:  # Regular case
        return verb + 'ing'

def check_lemma_in_category(word, category_word):
    # Process the words using spaCy to obtain lemmas
    doc_word = nlp(word.lower())
    word_lemma = doc_word[0].lemma_
    # Check if the lemma of the word exists in the lemmatized form of category_word
    #category_lemmas = [token.lemma_ for token in doc_category]
    return word_lemma in category_word

def sample_closest_from_category(word, category_name, samples_num=5):
    # Get all synsets for the specified category
    category_synsets = [synset for synset in wn.all_synsets() if synset.lexname() == category_name]
    # Extract words (lemmas) from the synsets
    category_words = set()
    for synset in category_synsets:
        for lemma in synset.lemmas():
            category_words.add(lemma.name())
    category_words = list(category_words)

    word_embedding = get_word_embedding(word)

    # Calculate similarity between word and each word in the category
    similarities = []
    for category_word in category_words:
      #word.lower() not in category_word.lower(): instead of checking whetherh the word itself exists in category_word.lower(), check whether the words' baseform exists in category_word.lower()
      #for example, if the word is reading, check whether read, reads... exists in category_word
        if not are_synonyms(word, category_word) and word.lower() not in category_word.lower():  # Ensure they are not synonyms
            category_embedding = get_word_embedding(category_word)
            similarity = cosine_similarity(word_embedding, category_embedding)
            similarities.append((category_word, similarity))

    # Sort the words by similarity and return the top N closest words
    similarities.sort(key=lambda x: x[1], reverse=True)
    optional = [word for word, similarity in similarities[:samples_num+10]]
    #print(optional)
    passed = [opt for opt in optional if not check_lemma_in_category(word, opt)][:samples_num]
    return passed



import random

def random_sample_from_predefined_category(predefined_categories_dict, word, category, samples_num=5, avoid_words=None):
    # Seed for reproducibility
    random.seed(36)
    
    # Check if the category exists in the predefined dictionary
    if category in predefined_categories_dict:
        # Get the list of words for the given category
        category_words = predefined_categories_dict[category]

        # Filter out words that are in the avoid_words list
        if avoid_words is not None:
            category_words = [w for w in category_words if w not in avoid_words]
        
        # Ensure the sample size doesn't exceed the filtered list length
        if len(category_words) <= samples_num:
            return category_words
        
        # Randomly sample the specified number of words
        sampled_words = random.sample(category_words, samples_num)
        return sampled_words
    else:
        # Return an empty list if the category doesn't exist
        return []



```

```python
import re
def create_modified_sentences(word, sentence, sampled_sub_words):
    # Use a regular expression to find exact whole word matches, considering word boundaries
    # This also preserves the original word's case
    word_regex = r'\b' + re.escape(word) + r'\b'

    modified_sentences = []
    for sub_word in sampled_sub_words:
        # Replace the word with sub_word in the sentence
        modified_sentence = re.sub(word_regex, sub_word, sentence, flags=re.IGNORECASE)
        modified_sentences.append(modified_sentence)

    return modified_sentences

def ensure_grammatical(model,tokenizer,sentence):
    tokenized_sentence = tokenizer('gec: ' + sentence, max_length=128, truncation=True, padding='max_length', return_tensors='pt')
    corrected_sentence = tokenizer.decode(
        model.generate(
            input_ids = tokenized_sentence.input_ids,
            attention_mask = tokenized_sentence.attention_mask, 
            max_length=128,
            num_beams=5,
            early_stopping=True,
        )[0],
        skip_special_tokens=True, 
        clean_up_tokenization_spaces=True
    )
    return corrected_sentence # -> I like swimming.
```

```python
import spacy
from itertools import permutations, combinations
def randomly_select_and_swap(swapping_types_dict, avoid_key_list):
    random.seed(36)
    # Filter out avoid_key from eligible keys
    if avoid_key_list:
        eligible_keys = [key for key, values in swapping_types_dict.items() if len(values) >= 2 and key not in avoid_key_list]
    else:
        eligible_keys = [key for key, values in swapping_types_dict.items() if len(values) >= 2]
    if not eligible_keys:
        return {}
    selected_key = random.choice(eligible_keys)
    selected_values = random.sample(swapping_types_dict[selected_key], 2)
    new_swapping_types_dict = {selected_key: selected_values}
    return new_swapping_types_dict

nlp = spacy.load('en_core_web_sm')

def swap_components(sentence,avoid_key_list=None):
    #print("avoid_key_list",avoid_key_list)
    sym_middle_words= ['next to','and','across','against','along', 'around','beside', 'besides','by','like','near', 'opposite','unlike', 'versus','with']
    # Parse the sentence
    doc = nlp(sentence)

    swapping_types_dict = {#swapping_types_dict['noun_phrases']
        'nouns': [],
        'adjectives': [],
        'adverbs': [],
        'noun_phrases': [(np.text, np.start, np.end) for np in doc.noun_chunks],
        'verb_phrases': []
    }
    # Extract noun phrases and adjust to exclude "the"
    np_indices = set()
    noun_phrase_items = []
    for np in doc.noun_chunks:
        for i in range(np.start, np.end):
            np_indices.add(i)

    for token in doc:

        if token.pos_ == 'NOUN':# and token.i not in np_indices:
            swapping_types_dict['nouns'].append((token.text, token.i, token.i + 1))
        elif token.pos_ == 'ADJ':
            swapping_types_dict['adjectives'].append((token.text, token.i, token.i + 1))
        elif token.pos_ == 'ADV':
            swapping_types_dict['adverbs'].append((token.text, token.i, token.i + 1))
        if token.head == token and token.pos_ == 'VERB':
            verb_phrase = " ".join([child.text for child in token.subtree if child.pos_ != 'PUNCT' and not child.is_stop])
            verb_start = min(child.i for child in token.subtree if child.pos_ != 'PUNCT' and not child.is_stop)
            verb_end = max(child.i for child in token.subtree if child.pos_ != 'PUNCT' and not child.is_stop) + 1
            swapping_types_dict['verb_phrases'].append((verb_phrase, verb_start, verb_end))

    #RANDOMLY SELECT a swapping type 
    #print("swapping_types_dict before",swapping_types_dict)
    swapping_types_dict = randomly_select_and_swap(swapping_types_dict,avoid_key_list)
    #print("swapping_types_dict after",swapping_types_dict)
   
    def swap_and_generate(sentence, items, doc, is_nouns):
        items_sorted = sorted(items, key=lambda x: x[1])
        swapped = []
        
        for perm in permutations(items_sorted):
            if perm != tuple(items_sorted):
                valid_swap = True
                tokens = sentence.split()
                for (orig_text, orig_start, orig_end), (new_text, new_start, new_end) in zip(items_sorted, perm):
                    if is_nouns:
                        # Check for prepositions and conjunctions between the original end and new start
                        prepositions_and_conj_between = [token.text for token in doc if (token.pos_ == 'ADP' or token.pos_ == 'CCONJ') and orig_end <= token.i < new_start]
                        
                        sym_prepositions_between = [p for p in prepositions_and_conj_between if p in sym_middle_words]
                        
                        if len(sym_prepositions_between) > 0:  # Don't swap if there are symmetric prepositions
                            valid_swap = False
                            break

                    # Perform the swap for valid pairs
                    orig_slice = slice(orig_start, orig_end)
                    tokens[orig_slice] = new_text.split()
                
                if valid_swap:
                    swapped.append(' '.join(tokens))
        return swapped


    def swap_and_adjust_noun_phrases(items,doc):
        swapped_sentences = []
        tokens_list = [(token.text_with_ws, pos, pos + 1) for pos, token in enumerate(doc)]
        items_sorted = sorted(items, key=lambda x: x[1])

        if len(items_sorted) > 2: 
            for item_pair in combinations(items_sorted, 2):
                # Find prepositions between the noun phrase pair
                start, end = min(item_pair[0][1], item_pair[1][1]), max(item_pair[0][2], item_pair[1][2])
                prepositions_between = [token.text for token in doc[start:end] if token.pos_ == 'ADP' or token.pos_ =='CCONJ']
                sym_preps_between = [p for p in prepositions_between if p in sym_middle_words]
                # Only swap if no symmetric preposition is found between the pair
                if len(sym_preps_between)==0:
                    for perm in permutations(item_pair):
                        if perm != tuple(item_pair):
                            modified_tokens_for_this_perm = [
                                (perm[0][0], perm[1][1], perm[1][2]),
                                (perm[1][0], perm[0][1], perm[0][2])
                            ]
                            for tok, start_pos, end_pos in tokens_list:
                                if tok.strip() not in perm[0][0] + ' ' + perm[1][0]:
                                    modified_tokens_for_this_perm.append((tok, start_pos, end_pos))
                            modified_tokens_for_this_perm = sorted(modified_tokens_for_this_perm, key=lambda x: x[1])
                            sen = ' '.join([m for m, _, _ in modified_tokens_for_this_perm])
                            sen = sen.replace('  ', ' ').replace(' .', '.')
                            sen = sen[0].upper() + sen[1:]
                            swapped_sentences.append(sen)

        #!!!GPT HELP: ADAPT THE SYM PREPS CONDITION FILTERING ALSO TO THIS len=2 case
        elif len(items_sorted) == 2:
            start, end = items_sorted[0][1], items_sorted[1][2]
            # Find prepositions between the two items
            prepositions_between = [token.text for token in doc[start:end] if token.pos_ == 'ADP' or token.pos_ == 'CCONJ']
            sym_preps_between = [p for p in prepositions_between if p in sym_middle_words]

            # Only proceed if there are no symmetric prepositions between the items
            if len(sym_preps_between) == 0:
                for perm in permutations(items_sorted):
                    if perm != tuple(items_sorted):
                        (orig_text, orig_start, orig_end), (new_text, new_start, new_end) = perm
                        modified_tokens_for_this_perm = [
                            (orig_text, new_start, new_end),
                            (new_text, orig_start, orig_end)
                        ]
                        for tok, start_pos, end_pos in tokens_list:
                            if tok.strip() not in orig_text + ' ' + new_text:
                                modified_tokens_for_this_perm.append((tok, start_pos, end_pos))
                        modified_tokens_for_this_perm = sorted(modified_tokens_for_this_perm, key=lambda x: x[1])
                        sen = ' '.join([m for (m, _, _) in modified_tokens_for_this_perm])
                        sen = sen.replace('  ', ' ').replace(' .', '.')
                        sen = sen[0].upper() + sen[1:]
                        swapped_sentences.append(sen)
        else:
            pass
        return swapped_sentences

    results = []
    if 'noun_phrases' in swapping_types_dict.keys():
        noun_phrase_items = swapping_types_dict['noun_phrases']
        if len(noun_phrase_items) > 1:  # Ensure there's something to swap
            swapped_versions = swap_and_adjust_noun_phrases(noun_phrase_items,doc)
            results.extend(swapped_versions)
            #print(results)
    

    for key, items in swapping_types_dict.items():
        if key!='noun_phrases' and len(items) > 1:  # Ensure there's something to swap
            if key =='nouns':
                swapped_versions = swap_and_generate(sentence, items,doc,is_nouns = True)
            else:
                swapped_versions = swap_and_generate(sentence, items,doc,is_nouns = False)
            results.extend(swapped_versions)

    used_key = list(swapping_types_dict.keys())
    if used_key:
        return results,used_key[0]
    else:
        return None,None


```

```python
import spacy
from transformers import GPT2TokenizerFast, GPT2LMHeadModel
import openai

def special_negate_to_nonnegated(sentence):
    contractions_map = {
            "isn't": "is",
            "aren't": "are",
            "wasn't": "was",
            "weren't": "were",
            "doesn't": "does",
            "don't": "do",
            "didn't": "did",
            "can't": "can",
            "couldn't": "could",
            "won't": "will",
            "wouldn't": "would",
            "haven't": "have",
            "hasn't": "has",
            "hadn't": "had",
            "shouldn't": "should",
            "mightn't": "might",
            "mustn't": "must"
        }
        
        # Split the sentence into words
    words = sentence.split()
    # Look for the contraction and replace it
    for i, word in enumerate(words):
        if word in contractions_map:
            words[i] = contractions_map[word]  # Replace with the affirmative form
            break
    
    # Rebuild the sentence after negation removal
    return ' '.join(words)
def negate_sentence(sentence):
    # Handle contraction-based negations
    if "not" in sentence:  # negative original sentence
        sentence = sentence.replace('not', '')
        return sentence
    elif "n't" in sentence:  # negative original sentence
        #find the verb with n't and then replace that word with the non-negated version of that verb
        return special_negate_to_nonnegated(sentence)
    else:  # positive sentences
        doc = nlp(sentence)
        negated_sentence = []
        negated = False

        for token in doc:
            # Special case for "is", just add "not" after it
            if token.text.lower() == "is" and not negated:
                negated_sentence.append(token.text + " not")
                negated = True
            # If an auxiliary verb is found, add "not" appropriately
            elif token.tag_ in ['VBZ', 'VBP', 'MD', 'VBD'] and not negated:
                if token.tag_ == 'VBD' and token.dep_ == 'ROOT':
                    negated_sentence.append("did not")
                elif token.tag_ == 'VBZ' and token.dep_ == 'ROOT':
                    negated_sentence.append("does not")
                else:
                    negated_sentence.append(token.text)
                    negated_sentence.append("not")
                negated_sentence.append(token.lemma_ if token.tag_ == 'VBZ' or token.tag_ == 'VBD' else token.text)
                negated = True
            elif token.tag_ == 'VB' and not negated:
                negated_sentence.append("do not")
                negated_sentence.append(token.text)
                negated = True
            else:
                negated_sentence.append(token.text)

        # Join tokens with proper spacing
        return " ".join(negated_sentence).replace(' .','.')

```
